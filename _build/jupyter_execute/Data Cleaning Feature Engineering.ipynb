# Data Cleaning

```python
import pandas as pd
```

```python
#loading the dataset
games = pd.read_csv('data/games.csv')
games
```

```python
#dropping column 'Unnamed: 0' that is not useful for our purposes
games = games.drop(columns = 'Unnamed: 0')
games
```

```python
# Narrow down our dataset down to previous wins and losses as well as the associated plus minus count.
games = games[['TEAM_NAME', 'GAME_ID', 'GAME_DATE', 'MATCHUP', 'WL', 'PLUS_MINUS']]
games
```

```python
games.info()
```

```python
# We want to change the dtype of Game_date from an object Date Time format.
games['GAME_DATE'] = pd.to_datetime(games['GAME_DATE'])
```

```python
games['GAME_DATE']
```

```python
# Now we can sort by date from in ascending order, from oldest to most recent.
games = games.sort_values('GAME_DATE')
games
```

## Feature Engineering

```python
# Now we will add features that will help in building our predictive model.
# We want to measure how well each team preformed before the current game.
# We will add the average plus minus score for the last 10 games for each team 
games['AVG_PLUS_MINUS_LAST10'] = games.groupby('TEAM_NAME')['PLUS_MINUS'].transform(lambda x: x.rolling(10, closed='left').mean())
games
```

```python
games[games['TEAM_NAME']=='Los Angeles Lakers'].head(35)
```

```python
# Now we would like to consider home court advantage. 
# We have to seperate the home and away games to be able to analyze this so we use the MATCHUP column to organize this. 
here_or_there = games['MATCHUP'].str.contains('@')
AWAY_GAMES = games[here_or_there]
HOME_GAMES = games[~here_or_there]
```

```python
HOME_GAMES.shape
```

```python
AWAY_GAMES.shape
```

```python
HOME_GAMES
```

```python
AWAY_GAMES
```

```python
# Now since we have so many duplicates because one team is home and one is away we would like to combine these for a cleaner data set to use in our prediction model.
games_merged = pd.merge(HOME_GAMES, AWAY_GAMES, on='GAME_ID', suffixes=('_HOME', '_AWAY'))
games_merged
```

```python
# Now we add another column for the avg plus minus difference between home and away 
games_merged['AVG_10_PLUS_MINUS_DIFF'] = games_merged['AVG_PLUS_MINUS_LAST10_HOME'] - games_merged['AVG_PLUS_MINUS_LAST10_AWAY']
```

```python
games_merged
```

```python
games_merged[['WL_HOME', 'AVG_10_PLUS_MINUS_DIFF']]
```

```python
model = games_merged[['WL_HOME', 'AVG_10_PLUS_MINUS_DIFF']].dropna()
```

```python
model
```

```python
# Now we have a model dataframe that we would use for our prediciton regression but first we need to change the Wins and Losses to 1s and 0. 
model['WL_HOME'] = model['WL_HOME'].map({'W': 1, 'L': 0})
model
```

```python
model.to_csv("games_model.csv")
```
