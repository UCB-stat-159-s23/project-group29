## Data Collection

We must install the NBA_API in order to collect our game data for our model. The API scrubs data from stats.nba.com.

```python
# First we must install the API
%pip install nba_api
```

```python
# Now we must import the league game finder class that will allow us to query for games.
from nba_api.stats.endpoints import leaguegamefinder
import numpy as np
# We create a class object called gamefinder that helps us find games and create a dataframe from the games we want to use.
gamefinder = leaguegamefinder.LeagueGameFinder(date_from_nullable='10/22/2019', date_to_nullable = '05/08/2023', league_id_nullable='00')
```

```python
# Using our API class object we can generate our first dataframe
games = gamefinder.get_data_frames()[0]
games.head()
```

```python
# Saving dataset to be stored in data folder for future reference.
games.to_csv('games.csv')
```
